```python
std_lot_area = StandardScaler()
std_year_built = StandardScaler()
std_year_remod = StandardScaler()
```

```python
df['LotArea'] = std_lot_area.fit_transform(df[['LotArea']])
df['YearBuilt'] = std_year_built.fit_transform(df[['YearBuilt']])
df['YearRemodAdd'] = std_year_remod.fit_transform(df[['YearRemodAdd']])
```

```python
std_ttl_bsmt_sf = StandardScaler()
df['TotalBsmtSF'] = std_ttl_bsmt_sf.fit_transform(df[['TotalBsmtSF']])
```

```python
from sklearn.preprocessing import LabelEncoder
mszoning_enco = LabelEncoder()
bldg_type_enco = LabelEncoder()
lot_config_enco = LabelEncoder()
df['MSZoning_enco'] = mszoning_enco.fit_transform(df['MSZoning'])
df['BldgType_enco'] = bldg_type_enco.fit_transform(df['BldgType'])
df['LotConfig_enco'] = lot_config_enco.fit_transform(df['LotConfig'])
```

```python

df[['LotConfig','LotConfig_enco']].drop_duplicates()
```

```python
df[['MSZoning','MSZoning_enco']].drop_duplicates()
```

```python
features = [
    'MSSubClass',
    'MSZoning_enco',
    'LotArea',
    'LotConfig_enco',
    'BldgType_enco',
    'OverallCond',
    'YearBuilt',
    'YearRemodAdd',
    'TotalBsmtSF'   
]
```

```python
y_true = df['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[features]
```

```python
model.fit(X,y_true)
```
